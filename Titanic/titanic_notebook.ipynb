```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('Data/train.csv')
test_data = pd.read_csv('Data/test.csv')
```

### Data Cleaning

```python
train_data.head()
```

```python
test_data.info()
```

```python
# Drop the Cabin column and the rows that have missing values
train_data = train_data.drop(['Cabin'], axis=1)
```

```python
train_data.info()
```

```python
# Convert categorical data to numerical data
train_data.loc[:, 'Sex'] = train_data['Sex'].apply(lambda x: 0 if x == "male" else 1)
train_data['Sex'] = train_data['Sex'].astype('int64')
```

```python
train_data.head()
```

```python
train_data["Embarked"].unique()
```

```python
train_data.loc[:, 'Embarked'] = train_data['Embarked'].apply(lambda x: 0 if x == "S" else (1 if x == "C" else 2))
train_data['Embarked'] = train_data['Embarked'].astype('int64')
```

```python
train_data.head()
```

```python
train_data["Age"].fillna(train_data["Age"].median(), inplace=True)
```

```python
train_data.info()
```

```python
train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']].corr()
```

### Prepare test data

```python
test_data.info()
```

```python
test_data = test_data.drop(['Cabin'], axis=1)
```

```python
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Embarked'].fillna(0, inplace=True)
```

```python
test_data.loc[:, 'Embarked'] = test_data['Embarked'].apply(lambda x: 0 if x == "S" else (1 if x == "C" else 2))
test_data.loc[:, 'Sex'] = test_data['Sex'].apply(lambda x: 0 if x == "male" else 1)
#convert sex to numeric
test_data['Sex'] = test_data['Sex'].astype('int64')
test_data['Embarked'] = test_data['Embarked'].astype('int64')
```

```python
test_data.info()
```

```python
test_data.head()
```

### Logistic Regression (Kaggle Score of 0.76315)

```python
import sklearn as sk
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
train_data.head()
```

```python
train_data = train_data.drop(['Name', 'Ticket', 'PassengerId'], axis=1)
```

```python
model = LogisticRegression(max_iter=1000)
x = train_data.drop(['Survived'], axis=1)
y = train_data['Survived']
model.fit(x, y)
```

```python
model.feature_names_in_
```

```python
test_data.head()
```

```python
test_X = test_data.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
test_data['Prediction'] = model.predict(test_X)
```

```python
test_data
```

```python
predictions = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': test_data['Prediction']})
```

```python
predictions.to_csv('predictions.csv', index=False)
```
